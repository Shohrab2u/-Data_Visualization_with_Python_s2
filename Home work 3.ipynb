## Create Two scatterplots with numeric columns

```python
plt.figure(figsize=(20,16))

plt.subplot(3, 1, 1)
sns.scatterplot(x="mileage", y="price", data=df)

plt.subplot(3, 1, 2)
sns.scatterplot(x="mpg", y="price", data=df)
plt.show()
```

```python
# import the data
# reading the csv file

input_file = 'telecom_data.csv'
telecom = pd.read_csv(input_file)
display(telecom.head())
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=telecom, x='Total_calls', y='Total_mins')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=telecom, x='Total_mins', y='Total_charges', hue='Churn')

plt.xlabel("total number of minutes")
plt.ylabel("total bills")
plt.title("scatterplot of total minutes vs total bill")

plt.xlim([0,900])
plt.ylim([0,500])
plt.grid()
plt.show()
```

## Create Two regression plots with numeric columns.

```python
plt.figure(figsize=(15,8))

# DICTIONARY = {key:value}
sns.regplot(data=telecom, x='Total_mins', y='Total_charges', line_kws={"color":"red"})

plt.xlabel("total number of minutes")
plt.ylabel("total bills")
plt.title("scatterplot of total minutes vs total bill")

plt.xlim([0,900])
plt.ylim([0,500])
plt.grid()
plt.show()
```

## Pair plot

```python
penguins = sns.load_dataset("penguins")

display(penguins.head())
print(penguins.shape)
```

```python
plt.figure(figsize=(20,16))

plt.subplot(2, 2, 1)
sns.regplot(x="mileage", y="price", data=df)

plt.subplot(2, 2, 2)
sns.regplot(x="mileage", y="price", data=df, line_kws={"color":"red"})

plt.subplot(2, 2, 3)
sns.regplot(x="mileage", y="price", data=df, scatter_kws={"color":"orange", "edgecolor":"white"})

plt.tight_layout()
plt.show()
```
